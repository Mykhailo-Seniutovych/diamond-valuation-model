```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ax = data.hist(bins=60, figsize=(20,10))
```

```python
plt.subplot(131), data['cut'].value_counts().reindex(['Ideal', 'Premium', 'Very Good', 'Good', 'Fair', 'Poor']).plot(kind='bar', figsize=(15,5))
plt.subplot(132), data['color'].value_counts().reindex(['D','E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'Z']).plot(kind='bar', figsize=(15,5))
plt.subplot(133), data['clarity'].value_counts().reindex(['FL', 'IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1', 'I2']).plot(kind='bar', figsize=(15,5))
plt.show()
```

```python
ax = data["carat"].hist(bins=50, figsize=(20, 10))
ax.set_xticks(np.arange(0.0, 6.0, 0.1))
plt.show()
outliers = data[data["carat"] > 3.0]
count = outliers.shape[0]
print("outliers_count", count)
```

```python
ax = data["depth"].hist(bins=50, figsize=(20, 10))
ax.set_xticks(np.arange(0, 100, 2))
plt.show()

outliers = data[np.logical_or(data["depth"] < 56, data["depth"] > 66)]
count = outliers.shape[0]
print("outliers_count", count)
```

```python
ax = data['table'].hist(bins=50, figsize=(20,10))
ax.set_xticks(np.arange(0, 100, 2))
plt.show()

outliers = data[np.logical_or(data["table"] < 50, data["table"] > 65)]
count = outliers.shape[0]
print("outliers_count", count)
```

```python
ax = data['price'].hist(bins=100, figsize=(20,10))
ax.set_xticks(np.arange(0, 20001, 1000))
plt.show()

outliers = data[np.logical_or(data["depth"] < 56, data["depth"] > 66)]
count = outliers.shape[0]
print("outliers_count", count)
```

```python
sns.pairplot(data,x_vars=['carat','cut','color','clarity'], y_vars='price', height=4)
sns.pairplot(data,x_vars=['depth','table','price'], y_vars='price', height=4)
```
